```python
# MultiGraph with attributes

G=nx.MultiGraph()
G.add_node('A',role='manager')
G.add_edge('A','B',weight= 6,  relation = 'friend', sign= '+')
G.add_edge('A','C',weight= 13,  relation = 'business partner', sign= '-')
G.add_edge('A','B',weight=2, relation = 'classmate')
G.node['A']['role'] = 'team member'
G.node['B']['role'] = 'engineer'

# draw the network
nx.draw_networkx(G)
```

```python
import networkx as nx
G= nx.MultiDiGraph()

G.add_edge('John', 'Ana', weight= 3, relation= 'siblings')
G.add_edge('Ana', 'David', weight= 4, relation= 'Cousins')
G.add_edge('Ana', 'Bob', weight= 1, relation= 'friends')
G.add_edge('Ana', 'Bob', weight= 1, relation= 'neighbors')
print( G.edge['Ana']['Bob'][1]['relation'])

# draw the network
nx.draw_networkx(G)
```
