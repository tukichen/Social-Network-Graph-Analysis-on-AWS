## Bipartite

```python
from networkx.algorithms import bipartite
B= nx.Graph()
B.add_nodes_from(['A','B','C','D','E'], bipartite= 0)  # label one set of notdes 0
B.add_nodes_from([1, 2, 3, 4], bipartite= 1) # label the other set of nodes 1
B.add_edges_from( [('A',1), ('B',1), ('C',1), ('C', 3), ('D', 2), ('E', 3), ('E',4)])

# draw the network
nx.draw_networkx(B)
```

```python
B.add_edge('A','B')
print(bipartite.is_bipartite(B))

B.remove_edge('A','B')
print(bipartite.is_bipartite(B))
```

```python
X = set([1,2,3,4])
print(bipartite.is_bipartite_node_set(B,X))
X = set(['A','B','C','D','E'])
print(bipartite.is_bipartite_node_set(B,X))
X = set(['A','B','C','D',1])
print(bipartite.is_bipartite_node_set(B,X))
```

### Projected Graphs

```python
X = set(['A','B','C','D','E'])
P = bipartite.projected_graph(B,X)
# draw the network
nx.draw_networkx(P)
```

```python
X = set([1,2,3,4])
P = bipartite.weighted_projected_graph(B,X)
# draw the network
nx.draw_networkx(P)
```
