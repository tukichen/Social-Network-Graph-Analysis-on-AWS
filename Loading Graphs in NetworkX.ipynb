# Loading Graphs in NetworkX

```python
import networkx as nx
import numpy as np
import pandas as pd
#%matplotlib notebook
%matplotlib inline
# initialge the graph
G1 = nx.Graph()
# add node/edge pairs
G1.add_edges_from([(0, 1),
                   (0, 2),
                   (0, 3),
                   (0, 5),
                   (1, 3),
                   (1, 6),
                   (3, 4),
                   (4, 5),
                   (4, 7),
                   (5, 8),
                   (8, 9)]
)
# draw the network
nx.draw_networkx(G1)
```

```python
# signed network

G2= nx.Graph()
G2.add_edge('A','B', sign ='+')
G2.add_edge('B','C', sign= '-')
# draw the network
nx.draw_networkx(G2)
```

```python
# MultiGraph with attributes

G=nx.MultiGraph()
G.add_node('A',role='manager')
G.add_edge('A','B',relation = 'friend')
G.add_edge('A','C', relation = 'business partner')
G.add_edge('A','B', relation = 'classmate')
G.node['A']['role'] = 'team member'
G.node['B']['role'] = 'engineer'

# draw the network
nx.draw_networkx(G)
```

```python
G = nx.dodecahedral_graph()
nx.draw(G)
nx.draw(G, pos=nx.spring_layout(G))  
```
